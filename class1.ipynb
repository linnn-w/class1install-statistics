```python
# plot a normal distribution curve 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
fig = plt.figure(figsize=(4,3))
x = np.linspace(-5, 5, 100) 
y = norm.pdf(x, 0, 1) # mean = 0, std = 1
plt.plot(x, y)
plt.title('Normal Distribution: mean = 0, std = 1')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```

```python
# plot 5 normal distribution curves with different means a
# and standard deviations
# 以下可以放或不放，因為已經在前面 import 過
# import numpy as np
# import matplotlib.pyplot as plt
from scipy.stats import norm

fig = plt.figure(figsize=(6,3))
x = np.linspace(-10, 10, 100)

# 用迴圈產生 mean 和 stds
for i in range(5):
    mean = i      # mean 從 0 到 4
    std =  1   # std 從 1 到 5
    y = norm.pdf(x, mean, std)
    plt.plot(x, y, label=f'mean={mean}, std={std}')

plt.title('Normal Distributions with Different Means and Stds')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
for i in range(5):
    mean = 0      # mean 從 0 到 4
    std =  i   # std 從 1 到 5
    y = norm.pdf(x, mean, std)
    plt.plot(x, y, label=f'mean={mean}, std={std}')    



plt.title('Normal Distributions with Different Means and Stds')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Microsoft JhengHei'  # 微軟正黑體
matplotlib.rcParams['axes.unicode_minus'] = False        
x = np.linspace(-10, 10, 100)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))  # 建立一行兩列的子圖

# 第一張圖：mean 變化，std 固定
for i in range(5):
    mean = i
    std = 1
    y = norm.pdf(x, mean, std)
    axs[0].plot(x, y, label=f'mean={mean}, std={std}')
axs[0].set_title('mean 變化, std=1')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Probability Density')
axs[0].legend()
axs[0].grid()

# 第二張圖：mean 固定，std 變化
for i in range(1, 6):
    mean = 0
    std = i
    y = norm.pdf(x, mean, std)
    axs[1].plot(x, y, label=f'mean={mean}, std={std}')
axs[1].set_title('mean=0, std 變化')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Probability Density')
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```
